```python
df['Faimly_size'] = df['SibSp'] + df['Parch'] + 1
```

```python
# Faimly type
# 1-> alone
# 2-4 -> small
# >5 -> large


def transform_family(num):
  if num==1:
    return 'Alone'
  elif num>1 and num<=4:
    return 'Small'
  else:
    return 'Large'
```

```python
df['Faimly_type'] = df['Faimly_size'].apply(transform_family)
```

```python
df[['Faimly_size','Faimly_type']]
```

```python
# bivariate analysis on new coloumns
pd.crosstab(df['Survived'],df['Faimly_type'],normalize='columns')*100
```
