### FINAL WORKING CODE (NumPy Linear Regression + Gradient Descent)


```python
import numpy as np

# --------------------
# 1. DATASET
# --------------------
X = np.array([1, 2, 3, 4, 5])
y = np.array([30, 40, 50, 60, 70])

# --------------------
# 2. PARAMETERS (START GUESS)
# --------------------
theta0 = 0
theta1 = 0

# --------------------
# 3. HYPOTHESIS (PREDICTION FUNCTION)
# --------------------
def predict(X, theta0, theta1):
    return theta0 + theta1 * X

# --------------------
# 4. COST FUNCTION (MSE)
# --------------------
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# --------------------
# 5. GRADIENT DESCENT TRAINING LOOP
# --------------------

learning_rate = 0.01
epochs = 1000
n = len(X)

for epoch in range(epochs):

    # Step 1: Predict
    y_pred = predict(X, theta0, theta1)

    # Step 2: Compute Cost
    cost = mse(y, y_pred)

    # Step 3: Compute Gradients
    d_theta0 = (-2/n) * np.sum(y - y_pred)
    d_theta1 = (-2/n) * np.sum(X * (y - y_pred))

    # Step 4: Update Parameters
    theta0 = theta0 - learning_rate * d_theta0
    theta1 = theta1 - learning_rate * d_theta1

    # Optional: print progress every 100 steps
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Cost = {cost}, theta0 = {theta0}, theta1 = {theta1}")

# --------------------
# 6. FINAL MODEL
# --------------------
print("\nFinal parameters:")
print("theta0:", theta0)
print("theta1:", theta1)

# --------------------
# 7. TEST PREDICTION
# --------------------
print("\nPrediction for 6 hours:")
print(theta0 + theta1 * 6)
```
