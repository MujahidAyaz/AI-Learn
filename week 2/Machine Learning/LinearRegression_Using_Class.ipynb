### What Even IS a Class?
A class is a blueprint. Like a blueprint for a house — the blueprint isn't a house, but you can build houses from it.

```python
# This is a BLUEPRINT (class)
class Dog:
    pass

# This is an ACTUAL DOG built from the blueprint
my_dog = Dog()
```

### Step 1 — Give It Data (__init__)
- __init__ runs automatically the moment you create the object. It sets up the starting data.

```python
class Dog:
    def __init__(self, name, age):  # runs automatically on creation
        self.name = name            # store name INSIDE the dog
        self.age  = age             # store age  INSIDE the dog

# Create two dogs from same blueprint
dog1 = Dog("Bruno", 3)
dog2 = Dog("Max", 5)

print(dog1.name)   # Bruno
print(dog2.name)   # Max
# self just means "this specific dog". Each dog has its own name and age.
```

### Step 2 — Give It Actions (Methods)
Methods are just functions inside a class.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def bark(self):                          # action
        print(f"{self.name} says: Woof!")

    def birthday(self):                      # another action
        self.age = self.age + 1              # changes the dog's own data
        print(f"{self.name} is now {self.age}")

dog1 = Dog("Bruno", 3)
dog1.bark()       # Bruno says: Woof!
dog1.birthday()   # Bruno is now 4
dog1.birthday()   # Bruno is now 5
```

### The Only 3 Rules to Remember
1.  self.something   →  data stored INSIDE the object
2.  def method(self) →  action the object can do
3.  __init__         →  runs automatically when object is created
